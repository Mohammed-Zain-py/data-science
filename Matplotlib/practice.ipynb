## One of the most common python library which is used for visualization in python is Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4,5])
plt.show()
```

# Putting axis values

```python
x = [1,2,3,4]
y = [4,5,6,7]

plt.plot(x,y)
```

# Giving title to the graph

```python
x = [1,2,3,4]
y = [4,5,6,7]

plt.plot(x,y)
plt.title("My first plot")
```

# Giving X and Y lables

```python
x = [1,2,3,4]
y = [80,90,95,98]

plt.plot(x,y)

plt.title("Marks corresponding to no. of hours studied")

plt.xlabel("no. of hours studied")
plt.ylabel("Marks Obtained")

plt.show()
```

# Showing only the written values : `xticks` and `yticks`

```python
x = [1,2,3,4]
y = [80,90,95,98]

plt.plot(x,y)

plt.title("Marks corresponding to no. of hours studied")

plt.xlabel("no. of hours studied")
plt.ylabel("Marks Obtained")

plt.xticks([1,2,3,4])
plt.yticks([80,90,98])

plt.show()

# The graph ddoesnot change, The x and y lable values change 
```

# Multiple Plots in the same figure

```python
x = [1,2,3,4]
dsa = [80,90,95,98]
ml_marks = [80,94,98,100]


plt.plot(x,dsa)
plt.plot(x,ml_marks)

plt.title("Marks corresponding to no. of hours studied")

plt.xlabel("no. of hours studied")
plt.ylabel("Marks Obtained")

plt.xticks([1,2,3,4])
plt.yticks([80,90,100])

plt.show()
```

### Now here the problem is that we dont know what the yellow line represents and what the blue line represents 

# So here comes the topic named : `legend`

```python
# same code as above

x = [1,2,3,4]
dsa = [80,90,95,98]
ml_marks = [80,94,98,100]


plt.plot(x,dsa, label = 'dsa marks')            # changed here
plt.plot(x,ml_marks, label = 'ml marks')        # changed here

plt.title("Marks corresponding to no. of hours studied")

plt.xlabel("no. of hours studied")
plt.ylabel("Marks Obtained")

plt.legend(loc="lower right")               # added this 

plt.xticks([1,2,3,4])
plt.yticks([80,90,100])

plt.show()
```

## Now we can do a lot of styling here 

#### how to know what all is available ? 
#
# Refer the documentation and scroll down

```python
print(plt.plot.__doc__)        # so here we see markers, colors, linewidth, linestyle and many more things lets try this in the above code
```

```python
# same code as above

x = [1,2,3,4]
dsa = [80,90,95,98]
ml_marks = [80,94,98,100]


plt.plot(x,dsa, label = 'dsa marks', color='green', marker = 'X',markersize = 10, linewidth = 2)            # changed here
plt.plot(x,ml_marks, label = 'ml marks')        

plt.title("Marks corresponding to no. of hours studied")

plt.xlabel("no. of hours studied")
plt.ylabel("Marks Obtained")

plt.legend(loc="lower right") 

plt.xticks([1,2,3,4])
plt.yticks([80,90,100])

plt.show()
```
