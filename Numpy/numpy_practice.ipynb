# Task 10 — Real mini problem

### Given:

##### prices = np.array([100, 200, 300, 400])
##### quantity = np.array([1, 2, 1, 3])

### Find:

##### total revenue per product
##### total revenue overall

```python
prices = np.array([100, 200, 300, 400])
quantity = np.array([1, 2, 1, 3])

total_revenue_per_product =  prices * quantity

total_revenue_overall = total_revenue_per_product.sum()

print(total_revenue_per_product)
print(total_revenue_overall)
```
