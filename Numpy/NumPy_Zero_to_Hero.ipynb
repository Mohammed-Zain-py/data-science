## 1. Basics
### 🧠 Theory
* **What is NumPy?** NumPy stands for Numerical Python. It is the core library for scientific computing in Python.
* **Why use NumPy?** It provides a high-performance multidimensional array object, and tools for working with these arrays. It is the foundation for pandas, scikit-learn, and basically all of ML in Python.
* **Difference between Python List and NumPy Array:**
  * NumPy arrays are **faster** because they are stored in continuous memory locations (C-type arrays).
  * NumPy arrays consume **less memory**.
  * NumPy arrays have a **fixed type** (e.g., all integers or all floats), while Python lists can contain mixed types.
  
Let's import NumPy. The standard convention is to import it as `np`.

## 2. Array Creation
### 🧠 Theory
You don't always need to create arrays from existing lists. NumPy provides built-in functions to quickly generate arrays of specific shapes and values.
* `np.array()`: Create from a list.
* `np.zeros()`: Array filled with 0s.
* `np.ones()`: Array filled with 1s.
* `np.full()`: Array filled with a specific value.
* `np.arange()`: Similar to Python's `range()`, creates a sequence.
* `np.linspace()`: Creates an array with evenly spaced values over a specified interval.
* `np.eye()`: Identity matrix (useful in Linear Algebra).

```python
# array() from list of lists (2D array)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", matrix)
```

```python
# zeros, ones, and full
zeros_arr = np.zeros((2, 3)) # 2 rows, 3 columns
ones_arr = np.ones((3, 2))
full_arr = np.full((2, 2), 99)

print("Zeros:\n", zeros_arr)
print("\nOnes:\n", ones_arr)
print("\nFull:\n", full_arr)
```

```python
# Sequence arrays
seq_arr = np.arange(0, 10, 2) # start=0, stop=10(exclusive), step=2
lin_arr = np.linspace(0, 1, 5) # 5 values evenly spaced between 0 and 1
eye_arr = np.eye(3) # 3x3 identity matrix

print("Arange:", seq_arr)
print("Linspace:", lin_arr)
print("Identity Matrix:\n", eye_arr)
```

## 3. Array Properties
### 🧠 Theory
Once you have an array, you'll need to check its structure, especially when debugging ML models.
* `ndim`: Number of dimensions (1D, 2D, 3D...).
* `shape`: The shape of the array (rows, columns).
* `size`: Total number of elements.
* `dtype`: Data type of the elements (e.g., int32, float64).
* `itemsize`: The size in bytes of each element.
* `reshape()`: Changes the shape of the array without changing its data.

```python
arr = np.array([[1.5, 2.5, 3.5], [4.5, 5.5, 6.5]])

print("Array:\n", arr)
print("Number of dimensions (ndim):", arr.ndim)
print("Shape (rows, cols):", arr.shape)
print("Total elements (size):", arr.size)
print("Data type (dtype):", arr.dtype)
print("Item size in bytes:", arr.itemsize)
```

```python
# Reshaping
flat_arr = np.arange(1, 13) # 1 to 12
reshaped_arr = flat_arr.reshape((3, 4)) # 3 rows, 4 columns

print("Original 1D:", flat_arr)
print("Reshaped 2D:\n", reshaped_arr)
```

## 4. Indexing and Slicing
### 🧠 Theory
Extracting specific values or chunks of an array.
* **1D Indexing:** Same as Python lists (`arr[0]`).
* **2D Indexing:** `arr[row, col]` (unlike lists which use `list[row][col]`).
* **Negative Indexing:** `-1` accesses the last element.
* **Slicing:** `arr[start:stop:step]`.

```python
arr_1d = np.array([10, 20, 30, 40, 50])
print("First element:", arr_1d[0])
print("Last element:", arr_1d[-1])
print("Slice [1:4]:", arr_1d[1:4])
print("Step slice [::2]:", arr_1d[::2]) # Every 2nd element
```

```python
arr_2d = np.array([[10, 20, 30], 
                   [40, 50, 60], 
                   [70, 80, 90]])

print("Element at row 1, col 2:", arr_2d[1, 2]) # 60
print("First row:", arr_2d[0, :])
print("First column:", arr_2d[:, 0])
print("Sub-matrix (top right 2x2):\n", arr_2d[0:2, 1:3])
```

## 5. Array Operations & Broadcasting
### 🧠 Theory
* **Element-wise operations**: In lists, `[1,2] + [3,4]` results in `[1,2,3,4]`. In NumPy, `np.array([1,2]) + np.array([3,4])` results in `[4,6]`. It does math!
* **Broadcasting**: NumPy's superpower! It allows math operations between arrays of different shapes (e.g., adding a scalar to a matrix).

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Power:", a ** 2)
print("Modulus:", b % 3)
```

```python
# Broadcasting example
matrix = np.array([[1, 2, 3], [4, 5, 6]])
scalar = 10
vector = np.array([100, 200, 300])

print("Matrix + Scalar:\n", matrix + scalar)
print("\nMatrix + Vector (Broadcast across rows):\n", matrix + vector)
```

## 6. Mathematical Functions
### 🧠 Theory
NumPy provides blazing-fast statistical and mathematical functions.
* `sum()`, `mean()`, `median()`, `min()`, `max()`
* `std()` (standard deviation), `var()` (variance)
* `sqrt()`, `exp()`, `log()`
You can apply these to the whole array, or along specific **axes** (axis 0 = columns, axis 1 = rows).

```python
arr = np.array([[1, 5, 3], 
                [4, 2, 6]])

print("Sum overall:", np.sum(arr))
print("Sum of each column (axis=0):", np.sum(arr, axis=0))
print("Sum of each row (axis=1):", np.sum(arr, axis=1))

print("\nMean:", np.mean(arr))
print("Max:", np.max(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# Universal Functions (ufuncs)
arr2 = np.array([1, 4, 9])
print("Square root:", np.sqrt(arr2))
print("Exponential (e^x):", np.exp(arr2))
```

## 7. Iteration
### 🧠 Theory
While you should avoid looping over NumPy arrays (use vectorization instead!), sometimes you have to.
* `for x in arr`: Loops over the first dimension.
* `np.nditer()`: Elegant way to loop through every single element of an n-dimensional array.

```python
arr2d = np.array([[1, 2], [3, 4]])

# Normal loop gives rows
print("Normal loop:")
for row in arr2d:
    print(row)

# nditer gives individual elements
print("\nnditer loop:")
for x in np.nditer(arr2d):
    print(x, end=" ")
```

## 8. Shape Manipulation (Reshape vs Resize vs Flatten vs Ravel)
### 🧠 Theory
* `reshape()`: Returns a new array with a new shape (doesn't modify original).
* `resize()`: Modifies the array *in-place*.
* `flatten()`: Returns a 1D *copy* of the array.
* `ravel()`: Returns a 1D *view* of the array (memory efficient, changes affect original).
* `transpose` or `.T`: Swaps rows and columns.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Original:\n", arr)
print("\nTransposed (.T):\n", arr.T)

flat = arr.flatten() # Copy
rav = arr.ravel()    # View
rav[0] = 999         # Changing ravel affects the original array!

print("\nAfter modifying ravel view, original array becomes:\n", arr)
```

## 9. Combining and Splitting Arrays

## 10. Boolean Masking & Filtering
### 🧠 Theory
This is the most common way to filter data in pandas and NumPy. You create a condition which returns an array of `True/False`, and use that as an index!
* `arr[arr > 5]`: Returns all elements greater than 5.
* `np.where(condition, x, y)`: Returns `x` if condition is true, else `y`.

```python
arr = np.array([1, 20, 15, 3, 50, 8, 12])

# Step 1: Create the boolean mask
mask = arr > 10
print("Boolean Mask:", mask)

# Step 2: Use mask to filter
print("Filtered Array (values > 10):", arr[mask])

# where() example: Replace values < 10 with 0, else keep original
new_arr = np.where(arr < 10, 0, arr)
print("Using np.where():", new_arr)
```

## 14. Real-world ML Relevance
### 🧠 Theory
How is this actually used?
1. **Feature Matrix ($X$) & Target Vector ($y$)**: ML datasets are loaded into 2D NumPy arrays ($X$) and 1D arrays ($y$).
2. **Min-Max Scaling / Normalization**: Bringing all features to a scale of 0 to 1 to help algorithms converge faster. Uses vectorized operations.

```python
# Dummy ML Feature Matrix X (3 samples, 2 features: age & salary)
X = np.array([
    [25, 50000],
    [35, 80000],
    [45, 120000]
])

# Normalization formula: X_scaled = (X - X_min) / (X_max - X_min)
# Notice how we specify axis=0 to find max/min for EACH column independently!
X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)

X_scaled = (X - X_min) / (X_max - X_min)

print("Original Data:\n", X)
print("\nNormalized Data (Values are between 0 and 1):\n", X_scaled)
```

## 15. Mini Project: Sales Performance Analysis
### 📊 Task:
You are given the sales data (in thousands of dollars) for 3 products over 4 quarters. 
Analyze the data using NumPy!

Data:
Product A: Q1=10, Q2=15, Q3=12, Q4=20
Product B: Q1=5, Q2=8, Q3=9, Q4=14
Product C: Q1=22, Q2=20, Q3=25, Q4=28

```python
# Setup
sales = np.array([
    [10, 15, 12, 20], # Prod A
    [5,  8,  9,  14], # Prod B
    [22, 20, 25, 28]  # Prod C
])

# 1. Total sales for each product across all quarters?
total_per_product = np.sum(sales, axis=1)
print("Total sales per product (A, B, C):", total_per_product)

# 2. Total sales across all products for each quarter?
total_per_quarter = np.sum(sales, axis=0)
print("Total sales per quarter (Q1, Q2, Q3, Q4):", total_per_quarter)

# 3. Find the maximum sales in a single quarter for Product C (Index 2)
max_prod_c = np.max(sales[2, :])
print("Max sales in a quarter for Product C:", max_prod_c)

# 4. Which product had the lowest overall total sales? (Hint: argmin)
worst_product_idx = np.argmin(total_per_product)
product_names = ["Product A", "Product B", "Product C"]
print("Worst performing product:", product_names[worst_product_idx])

# 5. Was there any quarter where sales for Product A exceeded 18k?
did_exceed = np.any(sales[0, :] > 18)
print("Did Product A ever exceed 18k?", did_exceed)
```

## 16. Placement / Interview Revision Notes 🏆

Before heading into your technical Data Science interview, review this rapid-fire list:

1. **Python List vs. NumPy Array:**
   - **Data**: Lists can contain mixed types; Arrays must be homogeneous (same type).
   - **Memory**: Arrays are densely packed in contiguous memory (C-style), lists are arrays of pointers. Arrays save memory.
   - **Speed**: Arrays are vastly faster due to Vectorization (operations happen in C, bypassing Python loops).

2. **`reshape` vs. `resize`:**
   - `reshape()`: Returns a *new* array with the new shape. The original array is unchanged.
   - `resize()`: Changes the shape of the array *in-place*. 

3. **`flatten` vs. `ravel`:**
   - `flatten()`: Returns a 1D **copy**. Modifying the flattened array will NOT change the original.
   - `ravel()`: Returns a 1D **view** (most of the time). Modifying it WILL change the original data. `ravel()` is faster and more memory efficient since it doesn't copy data.

4. **Vectorization & Broadcasting:**
   - **Vectorization**: Performing operations on entire arrays without writing explicit `for` loops in Python.
   - **Broadcasting**: A mechanism that allows NumPy to perform arithmetic on arrays of unequal shapes by implicitly "stretching/padding" the smaller array.
   
5. **Axis 0 vs Axis 1:**
   - By default, applying an aggregate function like `.sum()` squashes the whole array into a scalar.
   - Passing `axis=0` reduces the rows (giving you column totals).
   - Passing `axis=1` reduces the columns (giving you row totals).
