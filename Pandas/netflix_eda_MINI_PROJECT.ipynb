# Netflix Data Analysis using Pandas

This project performs Exploratory Data Analysis (EDA) on the Netflix titles dataset using Pandas.

The goal is to extract useful insights such as:
- content type distribution
- top countries
- release trends
- common ratings
- popular directors

This project marks the completion of my Pandas learning phase before moving to NumPy.

```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Dataset Overview

In this section, we inspect the structure of the dataset including:
- total rows and columns
- data types
- missing values

```python
df.info()
```

```python
df.isna().sum()
```

## Movies vs TV Shows

This analysis helps identify whether Netflix hosts more movies or TV shows.

```python
df['type'].value_counts()
```

## Top 10 Countries by Content Count

This shows which countries contribute the most content to Netflix.

```python
df['country'].value_counts().head(10)
```

## Top Release Years

This helps analyze the trend of content release across years.

```python
df['release_year'].value_counts().head(10)
```

## Latest Content Added

This section identifies the most recently released content available in the dataset.

```python
df[df['release_year'] == df['release_year'].max()]
```

## Most Common Content Rating

This helps identify the most frequent maturity rating in the catalog.

```python
df['rating'].mode()
```

## Top 10 Directors

This section highlights the directors with the most titles.

```python
df['director'].value_counts().head(10)
```

# Movies vs TV Shows percentage

```python
type_percentage = (df['type'].value_counts(normalize=True) * 100).round(2)
type_percentage
```

# Top 10 genres

```python
df['listed_in'].value_counts().head(10)
```

# Content growth after 2015

```python
df[df['release_year'] >= 2015]['release_year'].value_counts().sort_index()
```

# Movies released after 2020

```python
df[(df['type'] == 'Movie') & (df['release_year'] >= 2020)]
```

# Average release year by type

```python
df.groupby('type')['release_year'].mean()
```

# Top countries for TV Shows only

```python
df[df['type'] == 'TV Show']['country'].value_counts().head(10)
```

## Final Insights

- Movies dominate the Netflix catalog over TV Shows.
- The USA contributes the highest amount of content.
- Content releases increased significantly in recent years.
- Certain ratings dominate the platform's library.
- A few directors contribute multiple titles.
