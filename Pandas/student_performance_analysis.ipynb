```python
# 1. Read the CSV file into a DataFrame
import pandas as pd

df = pd.read_csv('students.csv')
```

```python
# 2. Show first 5 rows

df.head()
```

```python
# 3. Show dataset info

df.info()
```

```python
# 4. Show statistical summary

df.describe()
```

```python
# 5. Find student with highest Maths marks

highest_maths_marks = df.loc[df['Maths'].idxmax(), 'Name']
highest_maths_marks
```

```python
# 6. Find average attendance

avg_attendance = df['Attendance'].mean()
avg_attendance
```

```python
# 7. Show students with attendance > 90

df['Name'][df['Attendance']>90]
```

```python
# alternative approaches for the above:

df.loc[df['Attendance'] > 90, 'Name']    # using .loc
```

```python
df[df['Attendance'] > 90]      # This gives the full rows, not just names.
```

```python
# 8. Create a new column Average_Marks

df['Average_Marks'] = (df['Maths']+df['Science']+df['English'])/ 3
df
```

```python
# 9. Find topper based on average marks
topper = df.loc[df['Average_Marks'].idxmax(), 'Name']
topper
```

```python
# bonus :  if we want the details of the topper and not just the name then :

topper_details = df.loc[df['Average_Marks'].idxmax()]
topper_details
```
