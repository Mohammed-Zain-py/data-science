# Merging using joins

* Inner
* Outer
* Left
* Right

```python
df1 = pd.DataFrame({

    'id': [1,2,3,4],
    'names': ['Zain', 'Ali', 'Ankit', 'Juhi']
})

df2 = pd.DataFrame({

    'id': [3,4,5,6],
    'Course': ['ML', 'DL', 'Pandas', 'Python']
})
```

```python
print(df1)
print('---------------\n---------------')
print(df2)
```

```python
inner_result = pd.merge(df1,df2,how='inner',on='id')

inner_result   # only common columns 
```

```python
outer_result = pd.merge(df1,df2,how='outer',on='id')

outer_result   # all the columns 
```

```python
left_result = pd.merge(df1,df2,how='left',on='id')

left_result   # all columns from left and common columns from right
```

```python
right_result = pd.merge(df1,df2,how='right',on='id')

right_result   # all columns from right and common columns from left
```

# Concatination

```python
df3 = pd.DataFrame({

    'id': [1,2,3,4],
    'names': ['Zain', 'Ali', 'Ankit', 'Juhi']
})

df4 = pd.DataFrame({

    'id': [3,4,5,6],
    'Course': ['ML', 'DL', 'Pandas', 'Python']
})
```

```python
print(df3)
print('---------------\n---------------')
print(df4)
```

```python
result1 = pd.concat([df3,df4])
result1


# but as u cann see the index is not in correct order
```

```python
result2 = pd.concat([df3,df4], ignore_index=True)
result2

# now its proper but what if i want them side by side ?
```

```python
result3 = pd.concat([df3,df4], axis=1, )
result3

# now what if i want to know which values are from which dataframe ?
```

```python
result4 = pd.concat([df3,df4],keys=['df3', 'df4'])
result4
```
