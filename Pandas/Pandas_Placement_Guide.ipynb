## 10.  Handling Missing Values (fillna, dropna)
**Theory:** Real data is messy and often contains `NaN` (Not a Number) values.
- `dropna()`: Remove rows or columns containing missing values.
- `fillna()`: Replace missing values with specific data (e.g., mean, median, or fixed value).


```python
# Creating a dummy DataFrame with missing values
missing_data = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, np.nan, 90, np.nan],
    'Status': ['Pass', 'Fail', np.nan, 'Pass']
})
print("Original Missing Data:")
display(missing_data)

# dropna(): drop rows with ANY missing value
print("\nAfter dropna():")
display(missing_data.dropna())

# fillna(): fill Score with mean score, Status with 'Unknown'
print("\nAfter fillna():")
missing_data['Score'] = missing_data['Score'].fillna(missing_data['Score'].mean())
missing_data['Status'] = missing_data['Status'].fillna('Unknown')
display(missing_data)
```

## 12.  String Operations
**Theory:** Pandas provides the `.str` accessor to apply string functions (like upper, lower, contains, replace) element-wise on a column.

## 13.  Apply and Lambda
**Theory:** `.apply()` lets you apply a custom function to a row or column. Combined with `lambda` (anonymous functions), it becomes a highly flexible tool.


```python
# Create a salary bracket using apply and lambda
# Anyone > 80000 is High, else Medium
df_clean['Salary_Bracket'] = df_clean['Salary'].apply(lambda x: 'High' if x > 80000 else 'Medium')

display(df_clean[['Name', 'Salary', 'Salary_Bracket']])
```

## 14.  Groupby
**Theory:** Grouping involves splitting the data into groups, applying a function (like mean, sum, count), and combining the results. It's the core of descriptive data analytics!


```python
# Average salary per department
avg_salary = df_clean.groupby('Department')['Salary'].mean().reset_index()
print("--- Average Salary by Department ---")
display(avg_salary)

# Multiple aggregations at once
dept_stats = df_clean.groupby('Department').agg(
    Emp_Count=('Emp_ID', 'count'),
    Max_Exp=('Experience_Years', 'max'),
    Min_Salary=('Salary', 'min')
).reset_index()

print("\n--- Comprehensive Department Stats ---")
display(dept_stats)
```

## 15.  Merge and Concat
**Theory:**
- `concat`: Stacks DataFrames vertically (row-wise) or horizontally (column-wise).
- `merge`: Performs SQL-like joins mapping rows from two DataFrames based on common columns.


```python
# --- Concat Example (Vertical Stacking) ---
new_emps = pd.DataFrame({
    'Emp_ID': [109, 110], 
    'Name': ['Umar', 'Sara'], 
    'Department': ['HR', 'Data']
})

# Keep subset of columns for simplicity
df_concat = pd.concat([df_clean[['Emp_ID', 'Name', 'Department']], new_emps], ignore_index=True)
print("--- CONCATENATED ---")
display(df_concat.tail(4))

# --- Merge Example (Left Join) ---
bonus_data = pd.DataFrame({
    'Emp_ID': [101, 103, 109], 
    'Bonus_Amount': [5000, 4000, 2000]
})

df_merged = pd.merge(df_concat, bonus_data, on='Emp_ID', how='left')
print("\n--- MERGED (Left Join) ---")
display(df_merged.head(4))
display(df_merged.tail(2))
```

## 17.  Date-Time Operations

## 18.  Mini Project: App Store Dataset Analysis
**Task Scenario:** You have scraped a small dataset of mobile applications. You need to clean it and answer specific business questions.


```python
# 1. Load Dummy Dataset
app_data = {
    'App_Name': ['FitnessX', 'GameZ', 'FinancePro', 'StudyMate', 'GameZ', 'Health+', 'CryptoTrack'],
    'Category': ['Health', 'Gaming', 'Finance', 'Education', 'Gaming', 'Health', 'Finance'],
    'Rating': [4.5, np.nan, 4.8, 4.2, np.nan, 4.6, np.nan],
    'Downloads': ['10K', '500K', '50K', '20K', '500K', '100K', '5K'],
    'Price': ['$0.00', '$1.99', '$4.99', '$0.00', '$1.99', '$0.00', '$9.99']
}
apps = pd.DataFrame(app_data)
print("--- Original Raw Data ---")
display(apps)

# 2. Drop duplicates
apps = apps.drop_duplicates()

# 3. Handle Missing Ratings (Fill with average of available ratings)
avg_rating = apps['Rating'].mean()
apps['Rating'] = apps['Rating'].fillna(avg_rating).round(1)

# 4. Clean String Columns ('Downloads' to integer, 'Price' to float)
# Remove 'K' and multiply by 1000
apps['Downloads'] = apps['Downloads'].str.replace('K', '').astype(float) * 1000

# Remove '$' and convert to float
apps['Price'] = apps['Price'].str.replace('$', '').astype(float)

# 5. Question: Which category has the highest average downloads?
top_category = apps.groupby('Category')['Downloads'].mean().sort_values(ascending=False).head(1).reset_index()
print("\n--- Category with Highest Avg Downloads ---")
display(top_category)

# 6. Question: Add a column 'App_Type' -> 'Free' if Price == 0 else 'Paid'
apps['App_Type'] = apps['Price'].apply(lambda x: 'Free' if x == 0 else 'Paid')

print("\n--- Final Cleaned Data ---")
display(apps)
```
