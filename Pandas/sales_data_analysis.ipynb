```python
# 1. reading csv into DataFrame

import pandas as pd

df = pd.read_csv('sales.csv')
```

```python
# 2. showing first 5 rows

df.head()
```

```python
# 3. Creating the new column Total_Sales

df['Total_Sales'] = df['Price'] * df['Quantity']

df
```

```python
# 4. Find total sales of the entire dataset

df['Total_Sales'].agg(['sum'])
```

```python
# 5. Find the product with highest total sales

df.loc[df['Total_Sales'].idxmax(),'Product']
```

```python
# 6. Find all orders from Bangalore
df.loc[df['City'] == 'Bangalore']
```

```python
# 7. Find total sales city-wise using groupby()   (This is the most important task.)

df.groupby('City')['Total_Sales'].sum()
```

```python
# 8. Find category-wise average price

df.groupby('Category')['Price'].mean()
```

```python
# 9. Find the city with highest total sales

city_sales = df.groupby('City')['Total_Sales'].sum()
city_sales.idxmax()
```
