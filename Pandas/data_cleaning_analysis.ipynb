```python
# 1. Read CSV

import pandas as pd

df = pd.read_csv('employee_data.csv')
```

```python
# 2. Show first 5 rows

df.head()
```

```python
# 3. Check missing values in each column

df.isna().sum()
```

```python
# 4. Count total missing values in the entire dataset

df.isna().sum().sum()
```

```python
# 5. Show only rows having missing values

df[df.isna().any(axis=1)]
```

```python
# 6. Fill missing Salary values with average salary

df['Salary']=df['Salary'].fillna(df['Salary'].median())
df
```

```python
# 7. Fill missing City values with "Unknown"

df['City'] = df['City'].fillna('Unknown')
df
```

```python
# 8. Drop rows where Department is missing   (This simulates removing bad/incomplete records.)

df=df.dropna(subset=['Department'])
df
```

```python
# 9. Find average salary department-wise

df.groupby('Department')['Salary'].mean()
```
